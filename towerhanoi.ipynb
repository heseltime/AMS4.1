```python
import networkx as nx
```

```python
graph = nx.Graph()

# add nodes
for b1 in range(1, 4):
    for b2 in range(1, 4):
        for b3 in range(1, 4):
            graph.add_node(str(b1) + str(b2) + str(b3), data = [b1, b2, b3])

# add edges
for v, data in graph.nodes(data = "data"):
    for p in range(0, 3):
        old = data[p]
        if old > 0 and old in { data[i] for i in range(0, p) }:
            continue
        for new in range(1, 4):
            if new == old or new in { data[i] for i in range(0, p) }:
                continue
            target = v[:p] + str(new) + v[p+1:]
            graph.add_edge(v, target)
```

```python
nx.draw(graph, with_labels = True)
```
